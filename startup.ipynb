```python
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
endpoint = SPARQLWrapper('http://dbpedia.org/sparql')
query = """
    PREFIX ma: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>        
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    PREFIX dbpedia-owl: <http://dbpedia.org/ontology/>
    PREFIX dbpprop: <http://dbpedia.org/property/>
      
    CONSTRUCT {
        ?film rdf:type ma:Movie .
        ?film ma:title ?name .
        ?film ma:directed_by ?director .
        ?director rdf:type ma:Director .
        ?film ma:produced_by ?producer .
        ?producer rdf:type ma:Producer .
        ?film ma:features_actor ?actor .
        ?actor rdf:type ma:Actor .
        ?film ma:edited_by ?editor .
        ?editor rdf:type ma:Editor .
        ?film ma:cinematography_by ?cg .
        ?cg rdf:type ma:Cinematographer .
        ?film ma:distributed_by ?distributor .
        ?distributor rdf:type ma:Distributor .
        ?film ma:released_by ?studio .    
        ?studio rdf:type ma:Studio .
        ?film ma:has_production_country ?country .
        ?country rdf:type ma:Country .
        ?film ma:has_language ?lang .
        ?lang rdf:type dbpedia-owl:Language .
        ?film ma:budget ?budget .
        ?film ma:gross ?gross .
        ?film ma:plot_summary ?abstract .
        ?film ma:release_date ?date .
        ?film ma:runtime ?duration    
    }
    WHERE{
        ?film rdf:type dbpedia-owl:Film .
        ?film foaf:name ?name .

        OPTIONAL {?film dbpedia-owl:director ?director}
        OPTIONAL {?film dbpedia-owl:producer ?producer}
        OPTIONAL {?film dbpedia-owl:starring ?actor}
        OPTIONAL {?film dbpedia-owl:editing ?editor} 
        OPTIONAL {?film dbpedia-owl:cinematography ?cg}
        OPTIONAL {?film dbpedia-owl:distributor ?distributor}
        OPTIONAL {?film dbpedia-owl:studio ?studio}    
        OPTIONAL {?film dbpedia-owl:country ?country}
        OPTIONAL {?film dbpedia-owl:language ?lang}
        OPTIONAL {?film dbpprop:budget ?budget}
        OPTIONAL {?film dbpprop:gross ?gross} 
        OPTIONAL {?film dbpedia-owl:abstract ?abstract} 
        OPTIONAL {?film dbpedia-owl:releaseDate ?date}
        OPTIONAL {?film dbpprop:runtime ?duration}

        FILTER (LANG(?abstract)="en")
        FILTER regex(str(?name), "Harry Potter") 
    }"""
endpoint.setQuery(query)
endpoint.setReturnFormat(JSON)
```

```python
results = endpoint.query().convert()

for result in results['results']['bindings']:
  print(result)
```
